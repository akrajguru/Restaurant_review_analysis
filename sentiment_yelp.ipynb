```python
import numpy as np
import pandas as pd
from os import path
from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator

import matplotlib.pyplot as plt
```

```python
yelp_reviews = pd.read_csv('/content/drive/MyDrive/CS267_Project_Dataset/yelp_review.csv')
#yelp_reviews.head(10)
```

```python
# x = yelp_reviews['text'].values
# df['stars']=df['stars'].replace(to_replace=[1,2],value=0)
# df['stars']=df['stars'].replace(to_replace=[4,5],value=2)
# df['stars']=df['stars'].replace(to_replace=[3],value=1)
# y = df['stars'].values
import re
ratings = pd.DataFrame(yelp_reviews, columns=['business_id','stars','text','useful','funny','cool'])
ratings['text']=ratings['text'].apply(lambda x : re.sub('[^a-zA-Z0-9\s]',"",x))
ratings = ratings.iloc[:150]
#ratings.head(1)
```

```python
postive_df = ratings[ratings['stars'] > 3]
neutral_df = ratings[ratings['stars'] == 3]
negative_df = ratings[ratings['stars'] < 2]
```

```python
positive_text = " ".join(review for review in postive_df.text)
```

```python
wordcloud = WordCloud(width = 800, height = 800,
                background_color ='white',
                min_font_size = 10).generate(positive_text)
 
# plot the WordCloud image                      
plt.figure(figsize = (8, 8), facecolor = None)
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad = 0)
 
plt.show()
```

```python
negative_text = " ".join(review for review in negative_df.text)
wordcloud1 = WordCloud(width = 800, height = 800,
                background_color ='white',
                min_font_size = 10).generate(negative_text)
 
# plot the WordCloud image                      
plt.figure(figsize = (8, 8), facecolor = None)
plt.imshow(wordcloud1)
plt.axis("off")
plt.tight_layout(pad = 0)
 
plt.show()
```
